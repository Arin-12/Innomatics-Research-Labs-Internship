##Problem 1 — Mobile Recharge Validation System

This program validates a mobile recharge request.
Conditions:
• Minimum recharge amount ₹50
• Valid plans: 199, 299, 399, 599
• User can retry if invalid


```python
def recharge_validation():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge failed: Minimum recharge is ₹50")
            continue

        if amount in valid_plans:
            print("Recharge Successful! Plan Activated:", amount)
            break
        else:
            print("Invalid plan amount. Try again.")

# Run
recharge_validation()
```

##Problem 2 — Inventory Reorder Alert System

```python
def check_inventory(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(product, "→ Reorder Alert (Stock:", quantity, ")")
        else:
            print(product, "→ Stock OK (Stock:", quantity, ")")

# Sample Data
inventory = {
    "Keyboard": 10,
    "Mouse": 25,
    "Monitor": 8,
    "USB Cable": 40
}

check_inventory(inventory)
```

##Problem 3 — Student Result Processing System

```python
def student_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
marks = [60, 55, 40, 70, 65]
print("Result:", student_result(marks))
```

##Problem 4 — Cab Fare Estimator (with Retry)

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak.lower() == "yes":
        fare *= 1.25  # 25% extra

    return fare

while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    total = calculate_fare(distance, peak)
    print("Estimated Fare: ₹", round(total, 2))

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

##Problem 5 — Employee Attendance Eligibility Checker

```python
def attendance_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", round(percentage, 2), "%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
record = ["P", "P", "A", "P", "P", "A", "P", "P"]
print("Status:", attendance_eligibility(record))
```

##Problem 6 — Password Strength Checker

```python
def password_checker(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password (Too short)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Test
pwd = input("Enter password: ")
print(password_checker(pwd))
```
